```python
import matplotlib.pyplot as plt
labels = ['Random Forest', 'Decision Tree', 'Linear Regression', 'Multilayer Perceptron', 'Support Vector Regression', 'CNN', 'Bayesian Ridge', 'RF']
sizes = [210, 150, 180, 190, 210, 220, 130, 185]
# Plot
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```
