```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['R squared', 'RMSE', 'MAE']
RandomForest = [0.996, 17.563, 11.189]
LinearRegression = [0.993, 22.62, 17.63]
DecisionTree = [0.994, 21.353, 13.848]
MultilayerPerceptron = [0.994, 20.78, 14.28]
SupportVectorRegression = [0.992, 24.64, 14.75]
CNN = [0.995, 20.994, 14.63]
BayesianRidge = [0.993, 22.62, 17.63]
RF = [0.996, 17.586, 11.219]

x = np.arange(len(labels))  # the label locations
width = 0.10#he width of the bars

fig, ax = plt.subplots()
Random_Forest = ax.barh(x + width/3, RandomForest, width, label='RandomForest')
plt.axhline(x=0.996, color = 'r', linestyle = "dashed")
Linear_Regression = ax.barh(x + width/3, LinearRegression, width, label='LinearRegression')
plt.axhline(x=0.993,color = 'g', linestyle = "dashed")
Decision_Tree = ax.barh(x + width/3, DecisionTree, width, label='DecisionTree')
plt.axhline(x=0.994,color = 'g', linestyle = "dashed")
Multilayer_Perceptron = ax.barh(x + width/3, MultilayerPerceptron, label='MultilayerPerceptron')
plt.axhline(x=0.994,color = 'r', linestyle = "dashed")
Support_Vector_Regression = ax.barh(x + width/2, SupportVectorRegression, label='SupportVectorRegression')
plt.axhline(x=0.992,color = 'g', linestyle = "dashed")
CNN = ax.barh(x + width/3, CNN, label='CNN')
plt.axhline(x=0.995,color = 'r', linestyle = "dashed")


Bayesian_Ridge = ax.barh(x + width/2, BayesianRidge, label='BayesianRidge')
RF = ax.barh(x + width/3, RF, label='RF')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Regression Methods')
ax.set_ylabel('Values')
ax.set_title('Test Results')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(Random_Forest, padding=3)
ax.bar_label(Linear_Regression, padding=3)
ax.bar_label(Decision_Tree, padding=3)
ax.bar_label(Multilayer_Perceptron, padding=3)
ax.bar_label(Support_Vector_Regression, padding=3)
ax.bar_label(CNN, padding =3)
ax.bar_label(Bayesian_Ridge, padding=3)
ax.bar_label(RF, padding=3)

fig.tight_layout()

plt.show()
```
