```python
import matplotlib.pyplot as plt
   
SolarGridYear = [1980,1982,1990,1992,1993,1994,1999,2000,2010,2021]
GenerationAPU = [2.3,2.4,3.4,3.8,4.5,5.6,6.7,7.8,9.9,20.1]
  
plt.plot(SolarGridYear, GenerationAPU)
plt.title('SolarGridYear vs GenerationAPU')
plt.xlabel('SolarGridYear')
plt.ylabel('GenerationAPU')
plt.show()
```
